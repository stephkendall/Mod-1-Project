```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.the-numbers.com/person/39750401-Leonardo-DiCaprio#tab=acting')
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.findAll("table")
```

```python
movie_table = table[3]
```

```python
box_info = []
for table_row in movie_table.findAll('tr'):
    columns = table_row.findAll('td')
    all_info = []
    for column in columns:
        all_info.append(column.text)
    box_info.append(all_info)
```

```python
box_info = box_info[1:]
```

```python
df = pd.DataFrame(box_info, columns = ['Release Date', 'Title', 'Role', 'Domestic', 'International', 'Worldwide'])
```
